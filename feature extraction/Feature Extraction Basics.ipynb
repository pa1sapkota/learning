```python
image = imread('https://d27jswm5an3efw.cloudfront.net/app/uploads/2019/09/image-url-15.jpg') 
feature_matrix_image = np.zeros((534,800 ))
feature_matrix_image
```

```python
feature_matrix_image.shape
```

```python
image.shape
```

```python
# The Coloured Image has 3 dimension , so to get the average we use the looping 
for i in range(0, image.shape[0]): 
    for j in range(0, image.shape[1]): 
        feature_matrix_image[i][j] = ((int(image[i,j,0])+ int(image[i,j,1]) + int(image[i,j,0]) )/3 ) 
```

```python
feature_sample = np.reshape(feature_matrix_image, (534*800))
feature_sample
```
